```python
import numpy as np
import pandas as pd
from PIL import Image
from matplotlib import pyplot as plt
import math
```

```python
def pixel_distance(A,x,y,method):
    [m,n] = A.shape
    B = np.zeros((m,n), dtype=float)
    print("Matrix of ",m," by ",n)
   
    match method:
        case "D4": # City block distance
            print("City Block Method")
            for u in range(m):
                for v in range(n):
                    B[u,v] = abs(x-u) + abs(y-v)
            print(B)
            return B
        case "D8": # Chessboard distance
            print("Chessboard Method")
            for u in range(m):
                for v in range(n):
                    B[u,v] = max(abs(x-u) , abs(y-v))
            print(B)
            return B
        case "Eu":
            print("Euclidean Method")
            for u in range(m):
                for v in range(n):
                    B[u,v] = round(math.sqrt(pow((x-u),2) + pow((y-v),2)),2)
            print(B)
            return B
```

```python
m = 5
n = 5
A = np.zeros((m,n), dtype=float)
CityBlock = pixel_distance(A,0,0,"D4")
Chessboard = pixel_distance(A,0,0,"D8")
Euclidean = pixel_distance(A,0,0,"Eu")

# Show the results
plt.figure(0)
plt.imshow(CityBlock,cmap='gray')

plt.figure(1)
plt.imshow(Chessboard,cmap='gray')

plt.figure(2)
plt.imshow(Euclidean,cmap='gray')
```
